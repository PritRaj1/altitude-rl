```python
import numpy as np
from typing import Dict, Tuple
from tictactoe import TicTacToeMDP

env = TicTacToeMDP()
all_states = env.get_all_states()

# State-Action Value Function (Q-Table)
# Key: (state_tuple, action_tuple) -> float
Q_table_q = {}
```

# SARSA

SARSA (State-Action-Reward-State-Action) is an On-Policy Temporal Difference (TD) algorithm.

Unlike dynamic programming, which requires full knowledge of the environment, (transition probability $P(s^\prime \vert s,\pi(s))$ and reward $R(s,\pi(s),s^\prime)$), TD methods learn from sampled trajectories.

Unlike Q-Learning, which updates its values assuming perfect greedy play in the next state, SARSA updates its values based on the actual action chosen by its current policy ($\epsilon$-greedy) in the next state (on-policy).

This makes SARSA safer and more conservative, as it updates values factoring in immediate risks of potential mistakes.

$$
Q(S_t,A_t) \leftarrow Q(S_t,A_t)+\alpha\left[\underbrace{R_{t+1} + \gamma Q(S_{t+1}, A_{t+1})}_{\text{TD error}}−Q(S_t,A_t)\right]
$$

Where $\alpha$ is the learning rate. To learn the optimal value function $Q^*$ and optimal policy $\pi^*$, SARSA must gradually modify the policy $\pi$ so that it approaches optimality. This is achieved by making $\pi$ increasingly greedy with respect to the action-value estimates $Q$, similar to the dynamic programming policy improvement step.

However, making $\pi$ fully greedy and deterministic would violate the requirement that all state–action pairs be visited infinitely often. To avoid this, temporal-difference methods typically use an $\epsilon$-greedy policy, which balances exploration and exploitation, defined with parameter $\epsilon \in [0,1]$:

$$
\pi(a \mid s) =
\begin{cases}
1 - \epsilon + \frac{\epsilon}{|A(s)|}, & \text{if } a \in \arg\max_{a'} Q(s,a') \\
\frac{\epsilon}{|A(s)|}, & \text{otherwise}
\end{cases}
$$

```python
def train_sarsa(
    env: TicTacToeMDP,
    num_episodes: int = 50000,
    alpha: float = 0.1,
    gamma: float = 0.95,
    initial_epsilon: float = 1.0,
    min_epsilon: float = 0.05,
    decay_rate: float = 0.0001,
) -> Dict:

    Q = {}
    epsilon = initial_epsilon

    for episode in range(num_episodes):
        state = (0, 0, 1)
        is_terminal = False
        action = env.choose_epsilon_greedy_action(Q, state, epsilon)  # On policy step

        while not is_terminal and action is not None:
            next_state, reward, is_terminal = env.sample_environment_step(state, action)
            next_action = env.choose_epsilon_greedy_action(Q, next_state, epsilon)

            # SARSA TD Target
            next_q_val = 0.0
            if next_action is not None and not is_terminal:
                next_q_val = env.get_q_value(Q, next_state, next_action)

            td_target = reward + gamma * next_q_val

            # Policy Evaluation Step
            old_q = env.get_q_value(Q, state, action)
            Q[(state, action)] = old_q + alpha * (td_target - old_q)

            # Cycle forward
            state = next_state
            action = next_action

        epsilon = max(min_epsilon, epsilon * np.exp(-decay_rate))  # Decay learning rate
        if (episode + 1) % 10000 == 0:
            print(
                f"Episode {episode + 1}/{num_episodes} complete. Exploration Rate (Epsilon): {epsilon:.4f}"
            )

    return Q


Q_table_sarsa = train_sarsa(env, num_episodes=60000)
```

```python
# Extract target policy from final Q-Table
policy_sarsa = {}
for state in all_states:
    if not env.is_terminal(state):
        legal_actions = env.get_legal_actions(state)
        best_a = max(legal_actions, key=lambda a: env.get_q_value(Q_table_sarsa, state, a))
        policy_sarsa[state] = best_a

V_from_sarsa = {}
for state in all_states:
    if not env.is_terminal(state):
        legal_actions = env.get_legal_actions(state)
        V_from_sarsa[state] = max(
            env.get_q_value(Q_table_sarsa, state, a) for a in legal_actions
        )

print("\n--- SARSA AGENT VS STOCHASTIC ---")
env.play_simulation_game(policy=policy_sarsa, v_dict=V_from_sarsa, human_player=False)
```
