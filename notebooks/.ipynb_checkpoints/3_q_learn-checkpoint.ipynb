```python
import numpy as np
import random
from typing import Dict, Tuple, List
from tictactoe import TicTacToeMDP

env = TicTacToeMDP()
all_states = env.get_all_states()

# Initialize State-Action Value Function (Q-Table) 
# Key: (state_tuple, action_tuple) -> float
Q_table_q = {}
```

# Q-Learning 

Unlike SARSA, Q-learning is off-policy because its update rule assumes that from the next state $s^\prime$, the agent will select the absolute optimal action that maximises the current Q-table value, regardless of the actions the behavioral policy actually chooses. Meaning it samples from a different policy than its current.

$$
Q(S_t,A_t) \leftarrow Q(S_t,A_t)+\alpha \left[R_{t+1}+\gamma \max_a Q(S_{t+1},a)−Q(S_t,A_t)\right]
$$

```python
def train_q_learning(env: TicTacToeMDP, num_episodes: int = 50000, 
                     alpha: float = 0.1, gamma: float = 0.95, 
                     initial_epsilon: float = 1.0, min_epsilon: float = 0.05, 
                     decay_rate: float = 0.0001) -> Dict:
    
    Q = {}
    epsilon = initial_epsilon
    
    for episode in range(num_episodes):
        state = (0, 0, 1)
        is_terminal = False
        
        while not is_terminal:
            action = env.choose_epsilon_greedy_action(Q, state, epsilon)
            if action is None:
                break
                
            next_state, reward, is_terminal = env.sample_environment_step(state, action)
            
            # Off-Policy TD Target: max_a Q(s', a)
            next_legal = env.get_legal_actions(next_state)
            max_next_q = 0.0
            if next_legal and not is_terminal:
                max_next_q = max(env.get_q_value(Q, next_state, na) for na in next_legal)
                
            td_target = reward + gamma * max_next_q
            
            # TD Error Correction
            old_q = env.get_q_value(Q, state, action)
            Q[(state, action)] = old_q + alpha * (td_target - old_q)
            
            state = next_state
            
        epsilon = max(min_epsilon, epsilon * np.exp(-decay_rate)) # LR decay
        if (episode + 1) % 10000 == 0:
            print(f"Episode {episode + 1}/{num_episodes} complete. Exploration Rate (Epsilon): {epsilon:.4f}")
            
    return Q

Q_table_q = train_q_learning(env, num_episodes=60000)
```

```python
policy_q = {}
for state in all_states:
    if not env.is_terminal(state):
        legal_actions = env.get_legal_actions(state)
        best_a = max(legal_actions, key=lambda a: env.get_q_value(Q_table_q, state, a))
        policy_q[state] = best_a

V_from_q = {}
for state in all_states:
    if not env.is_terminal(state):
        legal_actions = env.get_legal_actions(state)
        V_from_q[state] = max(env.get_q_value(Q_table_q, state, a) for a in legal_actions)

print("\n--- Q-LEARNING AGENT VS STOCHASTIC ENVIRONMENT ---")
env.play_simulation_game(policy=policy_q, v_dict=V_from_q, human_player=False)
```
